Import necessary Python packages. Manual installation might be required for the packages below.

- [`matplotlib`](https://pypi.org/project/matplotlib/): data visualization and plotting

````python
from selenium import webdriver
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup

import pandas as pd
import matplotlib.pyplot as plt
````

Configure web driver for Chrome browser. You need to download Chrome Driver from the [official website](https://chromedriver.chromium.org/downloads).

**Note:** Each browser has its own web driver. For instance, Firefox browser and Safari browser (MacOS) have their own web drivers, which need to be downloaded and specified prior to the web scraping process.

## BeautifulSoup

Extract the content of the target page, then close the Chrome Driver.

````python
content = driver.page_source
soup = BeautifulSoup(content)

driver.close()
````

Extract information for all plants and their categories.

### Extract all HTML elements with `h3` and `table` tags

On the [Plants vs Zombies 2 - Plant page](https://plantsvszombies.fandom.com/wiki/Plants_(PvZ2)), the plant is displayed according to the world they are unlocked:

- the worlds are displayed as heading level 3 (HTML tag: `h3`)
- the plant tiles are displayed as table (HTML tag: `table`)

Therefore, we use `.find_all(<list of tags>)` method to extract all worlds and plants on the page **into a list** - note that further steps are required to preprocess extracted information as it might contain irrelevant results.

```
categories_and_plants = soup.find_all(['h3', 'table'])
```

### Initialize empty lists

Initialize empty lists to store plant details.

```
names = []

unlocks = []
effects = []
sun_costs = []
recharges = []

```

### Iterate over the list of all HTML elements

Loop through all the extracted information, keep only the relevant results.

```
for cp in categories_and_plants:
```

### Plant category

The category name (i.e., Player's House, Ancient Egypt, etc) are displayed as span that follows after the heading level 3, with class name "mw-headline" (`<h3><span class='mw-headline' ...>`). Therefore, we use `.find()` to extract the category name.

**Note:** `.find()` returns only the first result with the matching tag name and attributes, it is used here because there is only one span following the heading level 3 on the page.

```
    if cp.name == 'h3':
        cat_span = cp.find('span', {'class': 'mw-headline'})
        if cat_span:
            current_cat = cat_span.text
            print(f"\nCategory: {current_cat}\n{'-'*50}")
```

### Plant name and details

Each plant tile is a table on its own. The shared attributes between these tables are the style, i.e.: 

> `"text-align:left; width:98%; border: 0.5px solid black; cellpading:0; cellspacing:0;"`

The preprocessing step for each plant is split into two main parts:

**Part 1: Plant name**

Plant name is extracted from the left column of the plant tile, which is the link text below the plant image.

**Part 2: Plant details**

Plant details are extracted from the right column of the plant tile. The plant details starts with **bolded** text indicating each property (e.g., Description, Unlock, Plant Food effect, etc).

```

    elif cp.name == 'table':
```

````python
categories_and_plants = soup.find_all(['h3', 'table'])

names = []
categories = []
descriptions = []
unlocks = []
effects = []
sun_costs = []
recharges = []
sun_cost_remark = []
recharge_remark = []

extract_prop_details = lambda prop_name, full_str: full_str.replace(prop_name, '')

current_cat = ""
for cp in categories_and_plants:
    # Categories
    if cp.name == 'h3':
        cat_span = cp.find('span', {'class': 'mw-headline'})
        if cat_span:
            current_cat = cat_span.text
            print(f"\nCategory: {current_cat}\n{'-'*50}")
    
    # Plants
    elif cp.name == 'table':
        if cp.attrs['style'] == 'text-align:left; width:98%; border: 0.5px solid black; cellpading:0; cellspacing:0;':
            tds = cp.find_all('td', {'width': '20%'})
            
            # Plant name
            plant_name = ""
            for td in tds:
                if 'title' in td.find('a').attrs:
                    plant_name = td.find('a').text
                    
            names.append(plant_name)
            categories.append(current_cat)
            print(plant_name)
            
            # Plant properties
            plant_props = cp.find_all('b')
            for prop in cp.find_all('b'):
                # Plant description
                if 'Description' in prop.text:
                    descriptions.append(extract_prop_details('Description: ', prop.find_parent().text))
                    
                # Plant Food effect
                elif 'effect' in prop.text:
                    effects.append(extract_prop_details('Plant Food effect: ', prop.find_parent().text))
                    
                # Sun cost
                elif 'cost' in prop.text:
                    # 01 - Sun cost is fixed, can be converted to integer value.
                    try:
                        sun_costs.append(int(extract_prop_details('Sun cost: ', prop.find_parent().text)))
                        sun_cost_remark.append("")
                    # 02 - Sun cost is variable / not applicable, impute -1 into 'sun_cost' column, add 'sun_cost_remark'
                    except:
                        sun_costs.append(-1)
                        sun_cost_remark.append(extract_prop_details('Sun cost: ', prop.find_parent().text))
                
                # Recharge speed (in seconds)
                elif 'Recharge' in prop.text:
                    # 01 - Recharge speed is fixed, can be converted to float value.
                    try:
                        recharges.append(float(extract_prop_details('Recharge (in seconds): ', prop.find_parent().text)))
                        recharge_remark.append("")
                    # 02 - Recharge speed is not applicable, impute -1.0 into 'recharge', add 'recharge_remark'
                    except:
                        recharges.append(-1)
                        recharge_remark.append(extract_prop_details('Recharge (in seconds): ', prop.find_parent().text))
                
                # How to unlock plants (adventure, seed packet, money, gem)
                else:
                    unlocks.append(prop.find_parent().text)
            
            _ = [print(f"- {prop.find_parent().text}") for prop in plant_props]
            print()
````

## Pandas DataFrame

### Create Pandas DataFrame object

Create a Pandas DataFrame object from the extracted plant details.

````python
df = pd.DataFrame({
    'name': names,
    'category': categories,
    'description': descriptions,
    'unlock': unlocks,
    'effect': effects,
    'sun_cost': sun_costs,
    'recharge': recharges,
    'sun_cost_remark': sun_cost_remark,
    'recharge_remark': recharge_remark
})

display(df.head())
````

### Query Pandas DataFrame object

**Q:** Print the first 5 plants under **Power Mints** category.

````python
df[df['category']=='Power Mints'].head()
````

**Q:** Print all vine plants, i.e., plants which names contain "Vine". Sort the result by plant name in alphabetical order.

````python
df[df['name'].str.contains('Vine')].sort_values('name')
````

**Q:** Print all plants which sun cost is not less than 300. Sort the result first by sun cost in increasing order, then by plant name in reverse alphabetical order.

> **Observation:** All expensive plants have fast recharging rate (5 seconds each).

````python
df[df['sun_cost'] >= 300].sort_values(['sun_cost', 'name'], ascending=[True, False])
````

**Q:** Find all plants with zero sun cost, excluding all plants which belong to Power Mint category. Sort the result by plant name in alphabetical order. 

````python
df[(df['sun_cost'] == 0) & (df['name'].str.contains('mint', regex=False) == False)].sort_values('name')
````

**Q:** Plot histogram based on sun cost of the plants, excluding all plants with variable sun costs.

> **Observation:** Most plants cost 150 - 200 suns, whereas the minimum and maximum sun costs of all plants are 0 and 500 respectively.

````python
plt.hist(df[df['sun_cost']>=0]['sun_cost'])
````

**Q:** Plot histogram based on recharge rate of the plants, excluding all plants with variable recharge rate.

> **Observation:** Most plants fully recharge themselves within 5 - 20 seconds, some plants take more than 80 second to fully recharge themselves.

````python
plt.hist(df[df['recharge']>=0]['recharge'])
````

**Q:** Calculate average sun cost of the plants, excluding all plants with variable sun costs.

````python
df[df['sun_cost']>=0]['sun_cost'].mean()
````

**Q:** Find all plants which names contain any of these substrings: "So", "Su", "so", "su". Sort the result by plant name in alphabetical order.

````python
df[df['name'].str.contains('[Ss][ou]')].sort_values('name')
````

**Q:** Find all plants which names contain "Lettuce".

````python
df[df['name'].str.contains('Lettuce')]
````

**Q:** Find all plants which names contain "Z" or "z".

````python
df[df['name'].str.contains('[Zz]')]
````

**Q:** Find all plants with variable sun cost or recharge rate.

````python
df[(df['sun_cost'] == -1) | (df['recharge'] == -1)]
````
